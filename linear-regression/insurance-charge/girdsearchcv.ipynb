```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
df['species'].unique()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('species',axis=1)
y=df['species']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.33)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn=KNeighborsClassifier(n_neighbors=13)
```

```python
model_knn.fit(x_train,y_train)
```

```python
model_knn.score(x_test,y_test)
```

```python
from sklearn.svm import SVC
```

```python
model_svm=SVC(gamma='auto')
```

```python
model_svm.fit(x_train,y_train)
```

```python
model_svm.score(x_test,y_test)
```

```python
#now lets use gird search cv
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [1, 10, 20, 30],       # Note: C should be numbers, not strings
    'kernel': ['rbf', 'linear']
}
```

```python
classifier = GridSearchCV(
    estimator=model_svm,
    param_grid=param_grid,
    cv=5,
    return_train_score=True
)
```

```python
classifier.fit(x,y)
```

```python
results=pd.DataFrame(classifier.cv_results_)
```

```python
results
```

```python
results[['param_C','param_kernel','mean_test_score']]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
param_grid = {
    'C': [1, 10, 20, 30],
    'kernel': ['rbf', 'linear']
}
```

```python
classifier_2 = RandomizedSearchCV(
    estimator=model_svm,
    param_distributions=param_grid,
    cv=5,
    return_train_score=True,
    n_iter=4,
)
```

```python
classifier_2.fit(x,y)
```

```python
results=pd.DataFrame(classifier_2.cv_results_)
```

```python
results[['param_C','param_kernel','mean_test_score']]
```
